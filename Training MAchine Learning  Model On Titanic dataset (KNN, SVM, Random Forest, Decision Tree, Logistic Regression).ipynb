```python
file = pd.read_csv('titanic train.csv')
```

```python
file.Embarked.fillna('Q', inplace = True)
```

# Selecting Relevent Features

```python
feat = file[['Pclass', 'Sex', 'Age', 'SibSp', 'Embarked']]
label = file['Survived']
```

# Spliting Data

```python
xtrain, xtest, ytrain, ytest = train_test_split(feat, label, test_size = 0.20, shuffle = True)
```

# Training Model on KNN

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(xtrain, ytrain)
```

# Accuracy Score

```python
knn_score = knn.score(xtest, ytest)
```

# Prediction Plotting

```python
for i in xtrain:
    plt.scatter(xtrain[i], ytrain)
plt.plot(xtest, knn.predict(xtest), color = 'orange')
```

# Training Model on LogisticRegression

```python
log = LogisticRegression()
```

```python
log.fit(xtrain, ytrain)
```

# Accuracy Score
